## 13. 원 그래프(기본)

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) 
# autopct=.1f -> 소수점 첫번째 자리까지, %% -> %보이게
# startangle -> 시작점 지정(각도)
# counterclock=False -> 시계방향
plt.show()
```

```python
# 파이차트 조각 간격 띄우기

# explode = [0.2, 0.1, 0, 0, 0, 0]
explode = [0.1] * 6

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
# 파이차트 범례 설정
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3), title='언어별 선호도') # 범례의 제목 설정 가능
plt.show()
```

## 14. 원 그래프(심화)

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode = [0.05] * 6

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, explode=explode) 
plt.show()
```

```python
wedgeprops = {'width':0.7} #숫자가 작아질수록 얇아짐(0~1)
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, explode=explode, wedgeprops=wedgeprops) 
plt.show()
```

```python
wedgeprops = {'width':0.7, 'edgecolor':'w', 'linewidth':3} #숫자가 작아질수록 얇아짐(0~1)
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops) 
plt.show()
```

```python
# 파이차트에 글자 보여주기, 숨기기
def custom_autopct(pct):
    # return ('%.1f%%' % pct) if pct >= 10 else '' # 10% 이상인 것만 표시시켜주기
    # return '{:.1f}%'.format(pct) if pct >= 10 else '' # ↑위에꺼랑 같은 동작
    return '{:.0f}%'.format(pct) if pct >= 10 else '' # 정수로 표시하기

plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops, pctdistance=0.7) 
# pctdistance -> 글자가 중심에서 어느정도 떨어지는지, 클수록 멀어짐 (0~1)
plt.show()
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
# 점 크기 랜덤하게 설정
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
# 학년 기준으로 점 크기 다르게 만들기
sizes = df['학년'] * 500 # 1학년 500, 2학년 1000, 3학년 1500
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
# 학년별 점 색상 변경
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.5) # matplotlib cmap 검색하면 다양한 색상 종류 확인 가능
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize=(7, 7))

plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.5) # matplotlib cmap 검색하면 다양한 색상 종류 확인 가능
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5, orientation='horizontal')
# shrink -> colorbar 크기 조절
# orientation -> colorbar 위치? 조정
```

## 16. 여러 그래프

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10)) # 2 x 2에 해당하는 틀 생성
fig.suptitle('여러 그래프 넣기')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('여러 그래프 넣기')

# 첫 번째 그래프
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수') # 데이터 설정
axs[0, 0].set_title('첫 번째 그래프') # 제목
axs[0, 0].legend() # 범례
axs[0, 0].set(xlabel='이름', ylabel='점수') # x축, y축 label
axs[0, 0].set_facecolor('lightyellow') # 전경색 지정
axs[0, 0].grid(linestyle='--', linewidth=0.5)

# 두 번째 그래프
axs[0, 1].plot(df['이름'], df['수학'], label='수학')
axs[0, 1].plot(df['이름'], df['영어'], label='영어')
axs[0, 1].legend()

# 세 번째 그래프
axs[1, 0].barh(df['이름'], df['키'])

# 네 번째 그래프
axs[1, 1].plot(df['이름'], df['사회'], color='green', alpha=0.5)
```
