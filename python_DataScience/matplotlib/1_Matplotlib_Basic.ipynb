## 01. 그래프 기본
#### Matplotlib
- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

### 1. 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
```

```python
plt.plot(x,y)
plt.show() # plt.show()를 이용해서도 출력가능
```

#### Title 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프') # 한글 인코딩 문제로 ㅁㅁㅁ로 제목이 출력됨
```

#### 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 운영체제 사용할때
matplotlib.rcParams['font.size'] = 15 # 글자 크기
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist]
```

```python
# 한글 폰트 설정 후
plt.plot(x,y)
plt.title('꺾은선 그래프')
```

```python
plt.plot([-1, 0, 1], [-5, -1, 2]) # -가 인코딩 에러!
```

```python
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결

plt.plot([-1, 0, 1], [-5, -1, 2])
```

## 2. 축

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
plt.title('꺾은선 그래프')
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프', fontdict={'family':'HyGungSo-Bold', 'size':20}) # 개별로 폰트 설정
```

```python
# label 이름 추가
plt.plot(x, y)
plt.xlabel('X축')
plt.ylabel('y축')
```

```python
# label 색상 변경
plt.plot(x, y)
plt.xlabel('X축', color='red')
plt.ylabel('y축', color='#00aa00')
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red', loc='right') # left, center, right
plt.ylabel('y축', color='#00aa00', loc='top') # top, center, bottom
```

```python
# ticks 변경 하기
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```

## 3. 범례(legend)

```python
plt.plot(x,y, label='무슨 데이터')
plt.legend()
```

```python
plt.plot(x,y, label='무슨 데이터')
plt.legend(loc='upper right') # legend 위치 설정
```

```python
plt.plot(x,y, label='범례')
plt.legend(loc=(0.5, 0.5)) # x축, y축 (0~1 사이)
```

## 4. 스타일

```python
plt.plot(x, y, linewidth=10) # 그래프 라인 두께 설정
```

#### 마커(marker)

```python
plt.plot(x, y, marker='o') # 동그라미모양으로 마커
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='v') # 역삼각형 모양의 마커(matplotlib marker 검색하면 다양한 모양 확인)
```

```python
plt.plot(x, y, marker='v', markersize=10) # 마커 크기 변경
```

```python
plt.plot(x, y, marker='o', markersize=15, markeredgecolor='red') # 마커 테두리 색상 변경
```

```python
plt.plot(x, y, marker='o', markersize=15, markeredgecolor='red', markerfacecolor='yellow') # 마커 색 변경
```

#### 선 스타일

```python
plt.plot(x, y, linestyle=':') # 그래프 라인을 점선으로 표현
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.') # matplotlib linestyle 검색하면 더 다양한 모양의 linestyle 확인
```

#### 색깔

```python
plt.plot(x, y, color='pink')
```

```python
plt.plot(x, y, color='#ff0000') # RGB값도 입력 가능
```

```python
plt.plot(x, y, color='g') # g(green), b(blue)등 matplotlib color를 검색해서 기본으로 입력된 color 확인가능
```

#### 포맷

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:') # blue, 'v'(역삼각형), :(점선)
```

```python
plt.plot(x, y, 'go') # green, 'o'(동그라미), None -> 선이 나타나지 않음
```

#### 축약어

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':') 
# mfc -> markerfacecolor, ms -> markersize, markeredgecolor -> mec, ls -> linestyle
```

#### 투명도

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.3) # alpha -> 투명도 (0~1)
```

#### 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10,5), dpi=200) # dots per inch, 확대
plt.plot(x, y)
```

#### 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x,y)
```

## 6. 텍스트

```python
plt.plot(x, y, marker='o')

for idx, txt in enumerate(y):
    plt.text(x[idx], y[idx], txt)
```

```python
plt.plot(x, y, marker='o')

for idx, txt in enumerate(y):
    plt.text(x[idx], y[idx] + 0.3, txt)
```

```python
plt.plot(x, y, marker='o')

for idx, txt in enumerate(y):
    plt.text(x[idx], y[idx] + 0.3, txt, ha='center', color='blue')
```
