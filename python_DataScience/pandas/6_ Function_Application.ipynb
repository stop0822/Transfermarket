#### __데이터에 함수 적용(apply)__

```python
#'키'값 뒤에 cm를 붙히는 역할
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm) # 키 데이터에 대해서  add_cm 함수를 호출한 결과 데이터를 반영
df
```

```python
def capitalize(lang):
    if pd.notnull(lang): #NaN이 아닌지 확인
        return lang.capitalize() #첫 글자는 대문자, 나머지는 소문자로 변경
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```

```python
df['SW특기'].str.capitalize() #->위랑 같은 의미
```

## 14. 그룹화
- import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df동일한 값을 가진것들끼리 합쳐서 통계 또는 평균등의 값을 계산하기 위해 사용

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.groupby('학교')
```

```python
df.groupby('학교').get_group('북산고') #북산고 학생들만 추출
```

```python
df.groupby('학교').get_group('능남고') # 능남고 학생들만 추출
```

```python
df.groupby('학교').mean() # 계산 가능한 데이터들의 평균값
```

```python
df.groupby('학교').size() # 각 그룹의 크기
```

```python
df.groupby('학교').size()['능남고'] # 학교로 그룹화를 한 뒤에 능남고에 해당하는 데이터의 수
```

```python
df.groupby('학교')['키'].mean() # 각 학교의 평균 키 데이터 추출
```

```python
df.groupby('학교')[['국어', '영어', '수학']].mean() # 각 학교의 국어, 영어, 수학의 평균 데이터 추출
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2] # 학년 column 추가
df
```

```python
df.groupby(['학교', '학년']).mean() # 각 학교의 학년별 평균 데이터
```

```python
df.groupby('학년').mean() # 학년별 평균 데이터
```

```python
df.groupby('학년').mean().sort_values('키') # 학년별 키 평균데이터를 정렬하여 추출
```

```python
df.groupby(['학교', '학년']).sum()
```

```python
df.groupby('학교')[['이름', 'SW특기']].count() # 각 학교별 학생명, SW특기 데이터의 수(NaN값은 포함 X)
```

```python
school = df.groupby('학교')
school['학년'].value_counts() # 각 학교의 학년별 학생 수 추출
```

```python
school['학년'].value_counts().loc['북산고'] # 북산고의 학년별 학생 수 추출
```

```python
school['학년'].value_counts(normalize=True).loc['북산고'] # 북산고의 학년별 학생 수의 비율(%) 추출
```
