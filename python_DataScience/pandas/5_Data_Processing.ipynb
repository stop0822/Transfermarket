## 12. Data 수정

#### __Column 수정__

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['학교'].replace({'북산고':'상북고'}) # 학교column에 있는 '북산고'를 '상북고'로 값 변경
```

```python
df['학교'].replace({'북산고':'상북고', '능남고':'무슨고'}) # inplace=True를 해주어야 실제 데이터에 적용 가능
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True)
df
```

```python
df['SW특기'] = df['SW특기'].str.lower() # SW특기의 값들을 소문자로 변경
df
```

```python
df['SW특기'] = df['SW특기'].str.upper() # SW특기의 값들을 대문자로 변경
df
```

```python
df['학교'] = df['학교'] + '등학교' # '학교'데이터 값에 '등학교'추가 -> OOO + 등학교 = OOO등학교
df
```

#### __Column 추가__

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail' # '결과' column을 추가하고 전체 데이터는 Fail로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' # 총합이 400보다 큰 데이터에 대해서 결과를 Pass로 업데이트 시켜줌
df
```

#### __Column 삭제__

```python
df.drop(columns=['총합']) # '총합' column 삭제
```

```python
# 다중 선택
df.drop(columns=['국어', '영어', '수학']) # 국어, 영어, 수학 column 삭제
df # inplace=True를 줘야 실제 데이터 반영됨
```

#### __Row 삭제__

```python
df.drop(index='4번') # 4번 학생 데이터 row 삭제
```

```python
filt = df['수학'] < 80 # '수학'점수가 80점 미만 학생 필터링
df[filt]
```

```python
df[filt].index # '수학'점수가 80점 미만 학생의 index
```

```python
df.drop(index=df[filt].index) # '수학'점수가 80점 미만 학생의 index를 삭제
```

#### __Row 추가__

```python
df.loc['9번'] = ['이정환', '해남고등학교', 184, 90, 90, 90, 90, 90, 'Kotlin', 450, 'Pass'] # 새로운 Row 추가
df
```

#### __Cell 수정__

```python
df.loc['4번', 'SW특기'] = 'Python' # 4번 학생의 'SW특기'를 Python으로 값 변경
df
```

```python
df.loc['5번', ['학교', 'SW특기']] = ['능남고등학교', 'C']  # 5번 학생의 학교를 '능남고등학교', SW특기를 'C'로 값 변경
df
```
