## 8. Data 선택(iloc)
- 위치를 이용하여 원하는 row에서 원하는 column 선택

```python
df.iloc[0] # 0번째 위치의 데이터 추출
```

```python
df.iloc[4] # 4번째 위치의 데이터 추출
```

```python
df.iloc[0:5] # 0~4번째 위치의 데이터 추출
```

```python
df.iloc[0, 1] # 0번째 위치의 1번째(학교) 데이터 추출
```

```python
df.iloc[4, 2] # 5번 학생(4번째 위치)의 위치의 2번째(키) 데이터 추출
```

```python
# 다중선택
df.iloc[[0, 1], 2] # 0, 1번째 위치의 학생의 2번째(키) 데이터
```

```python
df.iloc[[0, 1], [3, 4]] # 0, 1번째 위치의 학생의 3, 4번째(국어, 영어) 데이터
```

```python
# 슬라이싱
df.iloc[0:5, 3:8] # 0~4번째 위치의 학생들의 3~7번째(국어~사회) 데이터 추출
```

## 9. Data 선택(조건)
- 조건에 해당하는 데이터 선택

```python
df['키'] >= 185 # 힉생들의 키가 185이상인지 여부를 boolean타입으로 확인
```

```python
filt = (df['키'] >= 185)
df[filt] # 키가 185이상인 학생들만 추출
```

```python
df[-filt] # 키가 185이상인 학생들을 제외한 학생들 추출
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185, '수학'] # 키가 185이상인 학생들의 수학점수 추출
```

#### __다양한 조건__
> &(and)

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')] # 키가 185이상인 북산고 학생 추출
```

#### __다양한 조건__
> |(or)

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)] # 키가 170미만이거나 200보다 큰 학생 추출
```

#### __str 함수__

```python
filt = df['이름'].str.startswith('송') # 송씨 성을 가진 사람 추출
df[filt]
```

```python
filt = df['이름'].str.contains('태') # '태'가 들어가는 사람 추출
df[filt]
```

```python
df[~filt] # '태'가 들어가는 사람을 제외한 사람 추출
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs) # SW특기가 Python이거나 Java인 사람
df[filt]
```

```python
df #황태산은 python이지만 대소문자로 인해 추출이 되지않음
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs) # SW특기들의 값을 소문자로 바꾸고(lower()) Python이거나 Java인 사람 추출
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java') # 4,5번 학생들의 SW특기 값이 NaN이므로 오류 발생
df[filt]
```

```python
df['SW특기'].str.contains('Java')
```

```python
df['SW특기'].str.contains('Java', na=True) # NaN 데이터에 대해 True로 설정
```

```python
df['SW특기'].str.contains('Java', na=False) # NaN 데이터에 대해 False로 설정
```

```python
filt = df['SW특기'].str.contains('Java', na=False) # SW특기에 'Java'를 포함하는 값을 가진 데이터 추출
df[filt]
```
